```python
# Conocer los maximos anotadores por partido PPG

best_PPG = df[['NAME','PPG']].sort_values('PPG', ascending=False)
best_PPG.head(10)
```

```python
# Conocer los que mas rebotes capturan

best_RPG = df[['NAME','RPG']].sort_values('RPG', ascending=False)
best_RPG.head(10)
```

```python
plt.figure(figsize = (16,2))
plt.bar(best_PPG['NAME'].head(10), best_PPG['PPG'].head(10))
```

```python
plt.bar(best_RPG["NAME"].head(10), best_RPG["RPG"].head(10))
```

```python
boston_players = []

for i, j in df[['NAME','TEAM']].values:
    if i not in boston_players and j == 'Bos':
        boston_players.append(i)

print(boston_players)
```

```python

PPG_dict = {}
for i, j in df[["NAME", "PPG"]].values:
    if i in boston_players:
        PPG_dict[i] = j

print(PPG_dict)
```

```python
plt.figsize = (10,10)

plt.bar(PPG_dict.keys(), PPG_dict.values())
plt.xticks(rotation = 90)
plt.show()
```
