```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Notas y Horas de estudio.csv')
```

```python
df.head(5)
```

```python
df.columns
```

```python
df.dropna(inplace=True)
```

```python
df.info()
```

```python
numerical_df = df.select_dtypes(include=['number'])
numerical_df.head()
```

```python
df_corr= numerical_df.corr()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
sns.heatmap(df_corr, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1)
plt.title('Matriz de Correlaci√≥n')
plt.show()
```

```python
sns.pairplot(numerical_df)
plt.show()
```

```python
for i in numerical_df.columns:
    sns.histplot(numerical_df[i], kde=True)
    plt.show()
```

```python
lista = ['Hours_Studied', 'Attendance', 'Parental_Involvement',
       'Access_to_Resources', 'Extracurricular_Activities', 'Sleep_Hours',
       'Previous_Scores', 'Motivation_Level', 'Internet_Access',
       'Tutoring_Sessions', 'Family_Income', 'Teacher_Quality', 'School_Type',
       'Peer_Influence', 'Physical_Activity', 'Learning_Disabilities',
       'Parental_Education_Level', 'Distance_from_Home', 'Gender' ]
for i in lista:
    sns.boxplot(data= df, x=i, y = 'Exam_Score')
    plt.show()
```
