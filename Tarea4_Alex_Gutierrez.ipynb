a. Un histograma para comparar una columna.

El análisis realizado en la tarea anterior, junto con la información reflejada en la matriz de correlación, revela que el contenido de alcohol es un factor crucial para determinar la calidad del vino. Este indicador muestra una fuerte relación con las características que definen un buen vino, sugiriendo que a medida que aumenta el porcentaje de alcohol, es más probable que el vino sea valorado con una mayor calidad. b. Un gráfico de burbujas para comparar 4 columnas.

Alcohol:

El contenido de alcohol es un factor crucial en la composición del vino.
Afecta directamente el sabor, cuerpo y percepción general del vino.
Niveles de alcohol más altos pueden indicar una mayor madurez de las uvas o un proceso de fermentación más largo.


pH:

El pH es una medida de la acidez del vino.
Influye significativamente en el sabor, la estabilidad microbiológica y el color del vino.
Un pH adecuado es esencial para la conservación y el envejecimiento del vino.


Sulfatos:

Los sulfatos son aditivos que actúan como conservantes y antioxidantes en el vino.
Ayudan a prevenir el crecimiento de levaduras y bacterias no deseadas.
La cantidad de sulfatos puede afectar el sabor y la estabilidad del vino.


Calidad:

Esta variable nos ayuda a representa una evaluación general de las características del vino.



Al visualizar estas variables juntas, se puede explorar:

Cómo el contenido de alcohol se relaciona con la calidad del vino.

Si hay una relación entre el pH y la calidad.

Cómo los niveles de sulfatos varían con respecto al alcohol y el pH.

Si vinos de mayor calidad tienden a agruparse en ciertas regiones del gráfico.

```python
plt.figure(figsize=(12, 8))
sns.scatterplot(data=df, x='alcohol', y='pH', size='sulphates', hue='quality', palette='viridis', sizes=(20, 200))
plt.title('Relación entre alcohol, pH, sulfatos y calidad')
plt.xlabel('Contenido de alcohol')
plt.ylabel('pH')
plt.show()
```

c. Un gráfico de dispersión para 2 columnas.

Elección de variables:

Alcohol: Es una característica fundamental del vino que afecta directamente su sabor, cuerpo y perfil organoléptico general.
Calidad: Es la variable objetivo, representando la evaluación general del vino.


Tipo de gráfico:
Un gráfico de dispersión es ideal para mostrar la relación entre dos variables numéricas continuas.

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='alcohol', y='quality')
plt.title('Relación entre contenido de alcohol y calidad')
plt.xlabel('Contenido de alcohol')
plt.ylabel('Calidad')
plt.show()
```

d. Una matriz de gráficos de dispersión para comparar un conjunto de 4 columnas.

Variables seleccionadas:

Alcohol: componente crucial que afecta el sabor y cuerpo del vino.

pH: indicador de acidez que influye en el sabor y la estabilidad.

Sulfatos: aditivos que actúan como conservantes y antioxidantes.

Calidad: la variable objetivo que representa la evaluación general del vino.

```python
sns.pairplot(df[['alcohol', 'pH', 'sulphates', 'quality']])
plt.suptitle('Matriz de dispersión para alcohol, pH, sulfatos y calidad', y=1.02)
plt.show()
```

e. Un gráfico de densidad para una columna, que separe los datos en vinos de buena calidad (calidad sobre 5) de mala calidad (bajo 5).

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(data=df[df['quality'] > 5], x='alcohol', shade=True, label='Buena calidad')
sns.kdeplot(data=df[df['quality'] <= 5], x='alcohol', shade=True, label='Mala calidad')
plt.title('Distribución del contenido de alcohol por calidad del vino')
plt.xlabel('Contenido de alcohol')
plt.ylabel('Densidad')
plt.legend()
plt.show()
```

f. Un gráfico de torta que muestre la cantidad de vinos de cada calidad.

En nuestro análisis de la calidad de los vinos, observamos que las calidades más comunes son 5 y 6. Además, contamos con un porcentaje pequeño de vinos de mayor calidad, como los de calidad 8. Es interesante notar que, aunque hay una presencia limitada de los peores vinos, su cantidad también es baja.

```python
quality_counts = df['quality'].value_counts()
plt.figure(figsize=(10, 8))
plt.pie(quality_counts.values, labels=quality_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribución de la calidad del vino')
plt.axis('equal')
plt.show()
```

g. Un small múltiples de histogramas, separados por calidad.

```python
g = sns.FacetGrid(df, col="quality", col_wrap=3, height=4, aspect=1.5)
g.map(sns.histplot, "alcohol", kde=True)
g.set_axis_labels("Contenido de alcohol", "Frecuencia")
g.set_titles("Calidad: {col_name}")
g.fig.suptitle('Distribución del contenido de alcohol por calidad del vino', y=1.02)
plt.show()
```

h. Un gráfico de violín donde se separe una columna en cuartiles y se grafique esta separación junto con la separación de vinos de buena y mala calidad (gráfico de violín con dos variables categóricas).

```python
df['alcohol_quartile'] = pd.qcut(df['alcohol'], q=4, labels=['Q1', 'Q2', 'Q3', 'Q4'])
df['quality_category'] = df['quality'].apply(lambda x: 'Buena' if x > 5 else 'Mala')

plt.figure(figsize=(12, 8))
sns.violinplot(data=df, x='alcohol_quartile', y='sulphates', hue='quality_category', split=True)
plt.title('Distribución de sulfatos por cuartil de alcohol y calidad del vino')
plt.xlabel('Cuartil de alcohol')
plt.ylabel('Sulfatos')
plt.show()
```

```python
import wikipedia
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import nltk
from wordcloud import WordCloud
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
import requests
from io import BytesIO

# Descargar recursos de NLTK si no están ya descargados
nltk.download('punkt')
nltk.download('stopwords')

# Obtener contenido de Wikipedia
page = wikipedia.page("Colombia",auto_suggest=False)
text = page.content

# Procesar el texto
stop_words = set(stopwords.words('spanish'))
word_tokens = word_tokenize(text.lower())
filtered_text = [w for w in word_tokens if not w in stop_words and w.isalpha()]

# Crear y mostrar la nube de palabras rectangular
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(' '.join(filtered_text))
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("Nube de palabras rectangular sobre Colombia")
plt.show()

# Obtener la imagen del sombrero colombiano
mask = np.array(Image.open('Sombrero.png'))

# Invertir la máscara para que el texto aparezca dentro del sombrero
mask = 255 - mask

# Crear y mostrar la nube de palabras con máscara de sombrero colombiano
wordcloud_masked = WordCloud(width=800, height=400, background_color='white', mask=mask, contour_width=3, contour_color='black').generate(' '.join(filtered_text))
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud_masked, interpolation='bilinear')
plt.axis("off")
plt.title("Nube de palabras con máscara de sombrero colombiano")
plt.show()
```
